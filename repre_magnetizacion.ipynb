```python
import matplotlib.pyplot as plt

# Nombres de los archivos para cada valor de N
archivos = {
    "N=32": "Mag32.txt",
    "N=64": "Mag64.txt",
    "N=128": "Mag128.txt"
}

# Crear la figura
plt.figure(figsize=(10, 6))

# Iterar sobre los archivos y graficar los datos
for label, archivo in archivos.items():
    temperaturas = []
    magnetizaciones = []
    errores = []

    # Leer los datos del archivo
    with open(archivo, "r") as file:
        for linea in file:
            magnetizacion, error, temperatura = linea.split()
            magnetizaciones.append(float(magnetizacion))
            errores.append(float(error))
            temperaturas.append(float(temperatura))

    # Graficar con barras de error
    plt.errorbar(temperaturas, magnetizaciones, yerr=errores, marker="o", linestyle="-", label=label)

# Configurar etiquetas y título
plt.xlabel("Temperatura", fontsize=12)
plt.ylabel("Magnetización Promedio", fontsize=12)
plt.title("Magnetización vs Temperatura para Distintos Valores de N", fontsize=14)
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

# Nombres de los archivos para cada valor de N (densidades positivas y negativas)
archivos_positivos = {
    "N=32 (Positiva)": "Dens32.txt",
    "N=64 (Positiva)": "Dens64.txt",
    "N=128 (Positiva)": "Dens128.txt"
}

archivos_negativos = {
    "N=32 (Negativa)": "DensNeg32.txt",
    "N=64 (Negativa)": "DensNeg64.txt",
    "N=128 (Negativa)": "DensNeg128.txt"
}

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar densidades positivas
for label, archivo in archivos_positivos.items():
    temperaturas = []
    densidades = []
    errores = []

    # Leer los datos del archivo
    with open(archivo, "r") as file:
        for linea in file:
            densidad, error, temperatura = linea.split()
            densidades.append(float(densidad))
            errores.append(float(error))
            temperaturas.append(float(temperatura))

    # Graficar con barras de error
    plt.errorbar(temperaturas, densidades, yerr=errores, marker="o", linestyle="-", label=label)

# Graficar densidades negativas
for label, archivo in archivos_negativos.items():
    temperaturas = []
    densidades = []
    errores = []

    # Leer los datos del archivo
    with open(archivo, "r") as file:
        for linea in file:
            densidad, error, temperatura = linea.split()
            densidades.append(float(densidad))
            errores.append(float(error))
            temperaturas.append(float(temperatura))

    # Graficar con barras de error
    plt.errorbar(temperaturas, densidades, yerr=errores, marker="x", linestyle="-", label=label)

# Configurar etiquetas y título
plt.xlabel("Temperatura", fontsize=12)
plt.ylabel("Densidad Media", fontsize=12)
plt.title("Densidad Media (Positiva y Negativa) vs Temperatura para Distintos Valores de N", fontsize=14)
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

# Nombre del archivo que contiene los datos
archivo = "energia_pmontecarlo.txt"

# Listas para almacenar los datos
pasos = []
energias = []

# Leer los datos del archivo
with open(archivo, "r") as file:
    for linea in file:
        energia, paso = linea.split()
        energias.append(float(energia))
        pasos.append(int(paso))

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(pasos, energias, marker="o", linestyle="", color="b", label="Energía promedio",markersize=2)

# Configurar etiquetas y título
plt.xlabel("Paso Monte Carlo", fontsize=12)
plt.ylabel("Energía Promedio", fontsize=12)
plt.title("Energía Promedio vs Paso Monte Carlo", fontsize=14)
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```
