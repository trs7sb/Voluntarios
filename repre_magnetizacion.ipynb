```python
import matplotlib.pyplot as plt

# Nombres de los archivos
archivo_superior = "magnetizacionsuperior.txt"
archivo_inferior = "magnetizacioninferior.txt"

# Listas para almacenar los datos
temperaturas = []
magnetizacion_superior = []
magnetizacion_inferior = []

# Leer el archivo de magnetización superior
with open(archivo_superior, "r") as file:
    for linea in file:
        magnetizacion, temperatura = linea.split()
        magnetizacion_superior.append(float(magnetizacion))
        temperaturas.append(float(temperatura))  # Las temperaturas son las mismas para ambos archivos

# Leer el archivo de magnetización inferior
with open(archivo_inferior, "r") as file:
    for linea in file:
        magnetizacion, _ = linea.split()  # La temperatura ya está en la lista
        magnetizacion_inferior.append(float(magnetizacion))

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(temperaturas, magnetizacion_superior, marker="o", linestyle=" ", color="b", label="Magnetización Superior")

# Configurar etiquetas y título
plt.xlabel("Temperatura", fontsize=12)
plt.ylabel("Magnetización Promedio", fontsize=12)
plt.title("Magnetización Superior e Inferior vs Temperatura", fontsize=14)
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```
